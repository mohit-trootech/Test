# Working with Remotes

To be able to collaborate on any Git project, you need to know how to manage your remote repositories. Remote repositories are versions of your project that are hosted on the Internet or network somewhere. You can have several of them, each of which generally is either read-only or read/write for you.

#### Create a Origin

As, we already progress much and know how to workflow a repo in local system now what we wanted to work in collaboration, with other devs, community we need a hosting platform.

Again, we created a one on github, now how to host this local repo on github just follow the commands.

1. go to github and copy the link or ssh key for your repo. *I recommend to use ssh as it is secure*



***ssh link***



***https link***

2. Now, open your git cli in same repo, and type this command 

```git remote add origin <link>```




3. Here we are we successfully made connection with github and local repos. now check the origin.

```git remote```

4. Check Remote repo links

```git remote -v```



5. Now We are ready to Push Changes to Github Repo.

**before pushing changes make sure to commit all the changes so safe and clean workflow**

```git push origin master```

*we used master because we don't have any branch if we have we have to target that branch for push changes*

```git push <remote-name> <branch-name>```
